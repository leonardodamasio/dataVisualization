```r
options(repr.plot.width=8, repr.plot.height=8, repr.plot.res=300)

split.screen(figs=c(2,2))

screen(1)
plot(trees$Girth, trees$Volume)

screen(2)
plot(trees$Girth, trees$Height)

screen(3)
plot(trees$Height, trees$Volume)

screen(4)
hist(trees$Volume)
```
